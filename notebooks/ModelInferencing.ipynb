```python
#Defining the model and providing the huggingface auth token
from transformers import AutoTokenizer, AutoModelForCausalLM, AutoModelForTableQuestionAnswering, AutoModelForSeq2SeqLM

model_name = "meta-llama/Llama-2-7b-hf"
auth_token = "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"  # Replace with your Hugging Face token


```

```python
#Text generation
# Load the tokenizer and model with the authorization token
tokenizer = AutoTokenizer.from_pretrained(model_name, use_auth_token=auth_token)
model = AutoModelForCausalLM.from_pretrained(model_name, use_auth_token=auth_token)

# Example usage: generating text
input_text = "Habari ya"
inputs = tokenizer(input_text, return_tensors="pt")
output = model.generate(**inputs, max_length=50)

# Decode and print the output
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)
```

```python
#Text summarization
# Load the tokenizer and model with the authorization token
tokenizer = AutoTokenizer.from_pretrained(model_name, use_auth_token=auth_token)
model = AutoModelForCausalLM.from_pretrained(model_name, use_auth_token=auth_token)

# Example text to summarize
input_text = ("Summarize the following text:"
              "Habari ya dunia! Habari ya asubuhi! Leo ni siku nzuri kwa kuanza miradi mipya na kuendeleza malengo yetu.")

# Tokenize the input text
inputs = tokenizer(input_text, return_tensors="pt")

# Generate the summary
outputs = model.generate(inputs.input_ids, max_length=100, num_return_sequences=1)

# Decode the output
summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("Summary:", summary)
```

```python
#Question Answering
from transformers import  AutoModelForQuestionAnswering
import torch
# Load the tokenizer and model with the authorization token
tokenizer = AutoTokenizer.from_pretrained(model_name, use_auth_token=auth_token)
model = AutoModelForQuestionAnswering.from_pretrained(model_name, use_auth_token=auth_token)

# Example context and question
context = """
Mji mkuu wa Kenya ni Nairobi. Huu ndio mji mkubwa zaidi nchini.
Inajulikana kwa urembo wake.
"""
question = "Mji mkuu wa Kenya ni upi?"

# Tokenize the input
inputs = tokenizer(question, context, return_tensors="pt")

# Get model predictions
with torch.no_grad():
    outputs = model(**inputs)

# Extract the answer
answer_start_index = torch.argmax(outputs.start_logits)
answer_end_index = torch.argmax(outputs.end_logits) + 1
answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(inputs.input_ids[0][answer_start_index:answer_end_index]))

print("Question:", question)
print("Answer:", answer)
```

```python
#Translation
# Load the tokenizer and model with the authorization token
tokenizer = AutoTokenizer.from_pretrained(model_name, use_auth_token=auth_token)
model = AutoModelForCausalLM.from_pretrained(model_name, use_auth_token=auth_token)

# Example text to translate
input_text = "Translate the following English text to Swahili: 'Hello, how are you?'"

# Tokenize the input text
inputs = tokenizer(input_text, return_tensors="pt")

# Generate the translation
outputs = model.generate(inputs.input_ids, max_length=100, num_return_sequences=1)

# Decode the output
translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("Translated Text:", translated_text)
```
